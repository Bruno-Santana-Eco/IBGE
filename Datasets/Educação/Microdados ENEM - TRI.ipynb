#### Renomear A,B,C em 1,2,3

```python
def criar_mapeamento(coluna):
    alternativas_unicas = sorted(coluna.unique())
    mapeamento = {alternativa: i + 1 for i, alternativa in enumerate(alternativas_unicas)}
    return mapeamento

# Aplicar a função a cada coluna do DataFrame
mapeamentos = {coluna: criar_mapeamento(df2[coluna]) for coluna in df2.columns}

# Aplicar os mapeamentos a cada elemento do DataFrame usando apply
df2 = df2.apply(lambda coluna: coluna.map(mapeamentos[coluna.name]) if coluna.name in mapeamentos else coluna)
df2
```

#### Unir os dataframes.

```python
df3['temp'] = df3.index
df3
```

```python
df2['temp'] = df2.index
df2
```

```python
# Unir usando a coluna temporária
df4 = pd.merge(df3, df2, on='temp', how='inner')
df4
```

```python
df4 = df4.drop('temp', axis=1)
df4
```
